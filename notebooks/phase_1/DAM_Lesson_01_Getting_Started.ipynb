## Exercise 4: Conditionals

Use `if/elif/else` to assign letter grades based on a numeric score.

```python
# Exercise 4: Conditionals
# Write a function called 'get_letter_grade' that:
#   - Takes a numeric score (0-100)
#   - Returns the letter grade:
#     90-100 = 'A', 80-89 = 'B', 70-79 = 'C', 60-69 = 'D', below 60 = 'F'

# TODO: your code here
def get_letter_grade(score):
    if score >= 90:
      return 'A'
    elif score >= 80:
      return 'B'
    elif score >= 70:
      return 'C'
    elif score >= 60:
      return 'D'
    else:
      return 'F'

# Test your function
print(get_letter_grade(95))  # Should print 'A'
print(get_letter_grade(83))  # Should print 'B'
print(get_letter_grade(72))  # Should print 'C'
print(get_letter_grade(65))  # Should print 'D'
print(get_letter_grade(45))  # Should print 'F'

    # Hint: use if/elif/else to check score ranges and return the letter
```

## Exercise 5: Grade Calculator Program
